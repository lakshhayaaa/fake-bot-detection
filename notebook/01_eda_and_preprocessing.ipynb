# Fake User / Bot Detection

## Objective
To analyze user data and preprocess it by selecting meaningful behavioral features for detecting bot accounts using machine learning.

## Observations

- Dataset contains both profile-based and behavioral features.
- Many behavioral features contain missing values.
- Target variable indicates whether an account is real (0) or bot (1).

## Missing Value Analysis

- Behavioral features such as posts_count, avg_likes, etc., have many missing values.
- These missing values indicate absence of activity data.
- Such rows are not useful for behavior-based bot detection.

## Feature Selection

We selected behavioral features because they reflect user activity patterns, which are harder to fake compared to profile attributes.

## Handling Missing Values

Rows with missing behavioral data were removed because they do not provide meaningful activity information for detecting bot behavior.
