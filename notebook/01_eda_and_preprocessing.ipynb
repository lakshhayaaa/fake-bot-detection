# BotHawk Dataset Preprocessing Pipeline

## Data Overview

The dataset contains ~19,778 samples with a strong class imbalance (~25:1 human to bot). No missing values were observed. Several numerical features exhibit heavy right-skew.


## Feature Selection

Non-informative features such as identifiers and low-variance attributes (login, bio, tag) were removed as they do not contribute meaningful predictive signal and may introduce noise.

## Outlier Handling

Numerical features showed extreme values due to heavy-tailed distributions. Winsorization at the 99th percentile was applied to cap extreme values while preserving data integrity.

## Skewness Handling

Several features exhibited high skewness (>2), indicating non-normal distributions. A conditional log transformation was applied to compress extreme values and improve distribution symmetry.

## Feature Engineering

Ratio-based features were created to capture relative behavioral patterns, such as activity intensity and social engagement, which are more informative than raw counts.

## Final Output

The resulting dataset is cleaned, statistically stable, and suitable for downstream modeling.
